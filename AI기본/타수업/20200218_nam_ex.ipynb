```python
#2월18일 수치미분 코드구현
#예제 5

import numpy as np

def func1(W):
    return W[0]**2

f = lambda W: func1(W)

def numerical_derivative(f,x): #x는 행렬, f는 함수
    delta_x = 1e-4
    grad = np.zeros_like(x) #저장할 미분값
    print("debug 1. initial input varivable = ", x)
    print("debug 2. initial grad = ", grad)
    
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
        
    while not it.finished:
        idx = it.multi_index
        
        print("debug 3. idx = ", idx, ", x[idx] = ", x[idx])
        
        tmp_val = x[idx] #numpy의 주의점것: 반드시 temp에 저장할 것
        x[idx]  = tmp_val + delta_x
        fx1 = f(x)
        
        x[idx] = float(tmp_val) - delta_x
        fx2 = f(x)
        
        grad[idx] = (fx1-fx2) / (2*delta_x)
        
        print("debug 4. grad[idx] =  ", grad[idx])
        print("debug 5. grad =  ", grad)
        
        x[idx] = tmp_val
        it.iternext()
        
    return grad


a = np.array([3.0])
ret1 = numerical_derivative(f, a)
print("type(ret) = ", type(ret1) , ", ret_val =" , ret1)

```

```python
#2월18일 수치미분 코드구현
#예제 6

import numpy as np

def func2(W):
    return (2*W[0]) + (3*W[0]*W[1]) +(W[1]**3)

f = lambda W: func2(W)

def numerical_derivative(f,x): #x는 행렬, f는 함수
    delta_x = 1e-4
    grad = np.zeros_like(x) #저장할 미분값
    print("debug 1. initial input varivable = ", x)
    print("debug 2. initial grad = ", grad)
    print("=======================================================")
    
    #it는 클래스형태
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    
    while not it.finished:
        idx = it.multi_index
        
        print("debug 3. idx = ", idx, ", x[idx] = ", x[idx])
        
        
        tmp_val = x[idx] #numpy의 주의점것: 반드시 temp에 저장할 것
        
        x[idx]  = float(tmp_val) + delta_x #x[idx] = x+dx
        fx1 = f(x) #f(x+dx)

        x[idx] = float(tmp_val) - delta_x  #x[idx] = x-dx
        fx2 = f(x) #f(x-dx)
        
        grad[idx] = (fx1-fx2) / (2*delta_x)
        
        print("debug 4. grad[idx] =  ", grad[idx])
        print("debug 5. grad =  ", grad)
        print("=======================================================")
        
        x[idx] = tmp_val #편미분을 위해(각각의 변수는 서로 독립적)
        it.iternext()
        
    return grad


a = np.array([1.0, 2.0])
ret2 = numerical_derivative(f, a)
print("type(ret2) = ", type(ret2) , ", ret_val =" , ret2)
```
