## Tidy Data 
- Tidy Data란 깔끔한 데이터 (=분석하기 좋은 데이터) 이다.
- Tidy Data의 3가지 조건 
    - 각 변수는 개별의 열(column)으로 존재한다.
    - 각 관측치는 행(row)를 구성한다.
    - 각 테이블은 단 하나의 관측기준에 의해서 조직된 데이터를 저장한다.
- Multi index처럼 복잡한 구조를 갖지 않는다.
- 모든 데이터를 DataFrame(2차원)을 기본으로 한다.

### join

`join`은 기본적으로 인덱스를 기준으로 데이터를 합친다.  
단, 이름이 겹치는 column이 있으면 Error를 발생시킨다.  
이때는 `lsuffix` 또는 `rsuffix` 옵션을 이용해서 컬럼명을 구분시켜줘야한다.

```python
# 같은 이름의 column이 있기 때문에 Error
a.join(b)
```

```python
# lsuffix는 왼쪽 데이터의 column명에 붙는 접미사
# rsuffix는 오른쪽 데이터의 column명에 붙는 접미사
a.join(b, lsuffix="_a", rsuffix="_b")
```

### merge

`merge`는 이름이 같은 column이 있으면, 해당 column을 기준으로 데이터를 합친다. (key 라고 한다.)  
또는 `on` 옵션을 이용해서 기준이 될 열(key)을 지정할 수도 있다.  
`join`과 달리 이름이 겹치는 열이 있어도 에러를 발생시키지 않는다.

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

left, right 모두 key 열을 가지고 있다. `merge`는 key 열을 기준으로 데이터를 합친다.

```python
result = pd.merge(left, right)
```

```python
result
```

## 실습: 영화 추천 알고리즘

이번 섹션의 실습 데이터는 LMS의 3월 11일 첨부자료에 있습니다. (movie-master.zip)

```python
user = pd.read_csv("movie/users.csv")
```

```python
user.head()
```

```python
movie = pd.read_csv("movie/movies.csv", encoding="cp949")
```

```python
movie.head()
```

```python
rating = pd.read_csv("movie/ratings.csv", encoding="cp949")
```

```python
rating.head()
```

`merge`로 데이터를 모두 합칠 수 있다.  
rating, user => "user_id" 열 공유
rating, move => "movie_id" 열 공유

```python
rating.merge(user).merge(movie)
```

추천에 불필요한 열을 먼저 제거한다.

```python
rating.drop(columns="timestamp", inplace=True)
```

user_id를 인덱스로, movie_id를 열로, 데이터는 rating으로 데이터를 가공한다.

```python
recommendation1 = rating.set_index(["user_id", "movie_id"]).unstack()
```

결측값(NaN)은 제거한다.

```python
recommendation1.fillna(0, inplace=True)
```

```python
recommendation1
```

`pivot`으로도 위와 같은 데이터를 만들 수 있다.

```python
rating.pivot("user_id", "movie_id", "rating").fillna(0)
```

데이터가 너무 많아 처리시간이 오래 걸리므로, 데이터를 줄여서 실습한다.

```python
recommendation2 = recommendation1.iloc[:1000, :1000].copy()
```

`corr`은 상관관계를 구하는 메서드이다. 

```python
recommendation2.corr()
```

```python
recommendation3 = recommendation2.T.corr()
```

```python
recommendation3
```

user_id가 310인 사용자와 비슷한 사람을 5명 뽑는다.

```python
recommendation3.loc[310].sort_values(ascending=False)[1:6] # 310번 자기 자신 제외
                                                           # 310번과 비슷한 상위 5명
```

310번과 취향이 가장 비슷한 사용자는 467번이므로, 467번이 본 영화는 310번도 좋아할 것이다.

```python
recommendation1.loc[467]
```

```python
seen467 = recommendation1.loc[467].sort_values(ascending=False)
```

```python
seen467[seen467 > 0].shape # 467번은 영화 69개 봄
```

467번이 본 영화 중에서 310번이 봤던 영화 제외하고 추천한다.

```python
seen310 = recommendation1.loc[310].sort_values(ascending=False)
```

```python
candidates = set(seen467[seen467 > 0].rating.index) - set(seen310[seen310 > 0].rating.index)
```

```python
len(candidates) # 추천 목록 39개
```
