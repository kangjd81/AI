## 기계학습

### 기계학습 핵심 고려사항

1. 데이터
2. 알고리즘/절차(Process)
3. 하드웨어(하드웨어 연동)

### 기계학습 분류

- 지도학습 (답이 있음)
- 비지도학습 (답이 없음)
- 강화학습 (환경)

### 지도학습

- 지도학습 : 과거 데이터를 기반으로 미래를 예측
    - classification (분류)
    - regression (회귀)

- 데이터는 많으면 많을 수록 좋다 (거의 모든 상황)  
- 데이터가 적을때, 데이터 전처리를 통해서 어느 정도 성능을 향상시킬 수 있다.
- 지도학습의 기본적인 가정: 우리가 수집한 데이터는 정답이다.
    - Bias(편향)가 없는 데이터
    - 잘못 측정된 데이터가 아님

- 예측 / 진단  
    - 숨어 있는 패턴을 찾는 것
    - 진단: 사람이 이해할 수 있는 것(White Box)
        - 의학계에서 필요
    - 예측: 사람이 이해할 수 없는 것(Black Box)
        - 데이터 마이닝/기계학습에 통용

## train_test_split

앞서 모델의 성능을 평가할 때는 학습에 사용했던 데이터를 성능 측정하는데 다시 사용했다.  
모델이 공부했던 데이터는 맞출 가능성이 높다. 그러므로 성능 측정이 공정하다고 할 수 없다.  
모델의 성능을 공정하게 측정하려면 모델이 한번도 본 적이 없는 데이터로 평가해야한다.  
이를 위해서 가지고 있는 데이터를 **학습용 데이터**와 **평가용 데이터**로 분리한다.  
이를 hold out 기법이라고 하며, scikit-learn에서는 `train_test_split` 함수로 구현되어 있다.

````python
from sklearn.model_selection import train_test_split
````

````python
train_test_split(breast.iloc[:, :-1], breast.iloc[:,-1])
````

앞이 컨테이너로 나오면 `len`으로 몇 개인가 체크해본다.

````python
len(train_test_split(breast.iloc[:, :-1], breast.iloc[:,-1]))
````

관례상 X_train, X_test, y_train, y_test 이라고 한다.

````python
X_train, X_test, y_train, y_test = train_test_split(breast.iloc[:, :-1], breast.iloc[:,-1])
````

## 성능을 올리기 위한 계속적인 노력  

*기계학습은 객관적 성능 측정 불가*

### 무한루프
성능을 올리는 과정은 다음 무한루프을 거친다.  
> 모델 생성 -> 성능체크 -> 수정 -> 모델 생성 -> 성능체크 -> 수정 -> 성능체크 -> ...
  
### 데이터에 대한 이해
- Domain Knowledge
    - 데이터에 관련된 전문 지식을 일컫는다.
- 성능을 올리려면 데이터를 더 잘 이해해야한다.
    - 결과를 해석하기 수월하다. 
    - 중복되거나 불필요한 feature를 제거한다.
    - feature를 조합해서 새로운 feature를 생성한다.
    - 결측치나 이상치가 발생하는 배경을 이해하고 값을 조정한다.
    - 기타 등
- Unicorn
    - 기계학습을 잘 하려면 1. 수학 & 통계  2. 프로그래밍  3. 전문지식(domain knowledge) 를 모두 잘해야한다.
    - 셋 다 잘하는 사람은 현실적으로 있을 수 없기 때문에 유니콘이라고 부른다.

## 열의 순서에 따라 성능 차이가 난다!(X)

열(Colunm) = 차원(Dimension) = (특성)Feature

````python
knn.fit(X_test_1, y_test_1)
````

````python
sum(knn.predict(X_test_1)==y_test_1)
````

두 모델의 성능은 같다.

````python
sum(knn.predict(X_test_2.values) == y_test_2)
````

````python
sum(knn2.predict(X_test_2.values[:, ::-1]) == y_test_2) # 예측할 때도 열 순서 거꾸로
````

## train_test_split 옵션

`train_test_split`의 Signature는 다음과 같다.  
가변 포지셔널 인자로 데이터를 받고, 가변 키워드 인자로 옵션을 받는다.  
옵션에 따라서 데이터를 쪼개는 방법이 다양하다.
```python
train_test_split(*arrays, **options)
```

`test_size`는 데이터를 쪼개는 비율을 바꿔준다.

`random_state`는 random seed와 같다.  
> ※ 참고  
컴퓨터 프로그램에서 발생하는 무작위 수는 사실 엄격한 의미의 무작위 수가 아니다. 어떤 특정한 시작 숫자를 정해 주면 컴퓨터가 정해진 알고리즘에 의해 마치 난수처럼 보이는 수열을 생성한다. 이런 시작 숫자를 시드(seed)라고 한다.  
출처: [데이터 사이언스 스쿨](https://datascienceschool.net/view-notebook/8bf41f87a08b4c44b307799577736a28/)

````python
a1, b1 = train_test_split(breast, test_size=0.5, train_size=0.5)
a2, b2 = train_test_split(breast, test_size=0.5, train_size=0.5)
print("a1과 a2는 같다: ", a1.equals(a2))
print("b1은 b2와 같다: ", b1.equals(b2))
````

````python
a1, b1 = train_test_split(breast, test_size=0.5, train_size=0.5, random_state=1)
a2, b2 = train_test_split(breast, test_size=0.5, train_size=0.5, random_state=1)
print("a1과 a2는 같다: ", a1.equals(a2))
print("b1은 b2와 같다: ", b1.equals(b2))
````
