### underfitting

- 로스가 거의 떨어지지 않는 경우는 학습이 안 되고 있는 것이다. 

- 일반적으로 train loss가 validation loss보다 항상 작다.  
- train loss는 학습에 사용했던 train data로 측정하는 것인데 한번도 안 본 validation data를 더 잘 맞추는 것이 이상하다.
- 딥러닝은 underfitting이 잘 생기지 않는다.

### overfitting

- train loss는 계속 줄어드는데 validation loss는 커지는 경우는 오버피팅이다. 
- validation loss가 높아지는 시점부터 overfitting이 발생했다고 본다.
- 예측의 편차가 크면 그래프가 요동친다. → overfitting
    - overfitting : High Variance, Low Bias
    - underfitting : Low Variance, High Bias

### good fitting

- train loss와 validation loss가 둘다 낮고 거의 비슷해진다.
- 이상적인 learning curve의 모양이다.

###  a Training Dataset That May Be too Small Relative to the Validation Dataset

- train data가 너무 적을 경우

### a Validation Dataset That May Be too Small Relative to the Training Dataset.

- validation data가 너무 작을 경우 validation loss가 요동칠 수 있다. 

- validation loss가 train loss보다 낮다. 
    - validation data가 너무 작으면 그럴 수 있다.

one hot encoding은 열을 여러개로 만들고, 1개만 1로 하고 나머지는 0으로 바꾸었다.  
multi hot encoding은 열을 여러개로 만들고, 1이 여러개 있고 나머지는 0으로 바꾼다.

````python
# 튜토리얼코드

def multi_hot_sequences(sequences, dimension): 
    # 0으로 채워진 (len(sequences), dimension) 크기의 행렬을 만듭니다
    results = np.zeros((len(sequences), dimension))
    for i, word_indices in enumerate(sequences):
        results[i, word_indices] = 1.0  # results[i]의 특정 인덱스만 1로 설정합니다
    return results
````

### [가중치를 규제하기](https://www.tensorflow.org/tutorials/keras/overfit_and_underfit?hl=ko#%EA%B0%80%EC%A4%91%EC%B9%98%EB%A5%BC_%EA%B7%9C%EC%A0%9C%ED%95%98%EA%B8%B0)

오버피팅을 막는 또 하나의 기법. `Dense`에는 `kernel_regularizer`라는 것이 있다.  
가중치의 크기를 규제하는 기법이다. weight decay(가중치 감쇠) 또는 정규화라고 부르기도 한다.  


<center>이미지 출처 : https://www.quora.com/When-would-you-chose-L1-norm-over-L2-norm </center>

regularization은 L1, L2 기법이 있다. L1은 마름모 안으로 가중치의 크기를 제한하고, L2는 원 안으로 가중치의 크기를 제한한다.   
L1과 L2를 혼합한 elastic-net 등도 있다.

````python
def create_model(unit):
    """인자에 따라서 레이어의 unit 수를 바꾸는 함수.
    첫번째 레이어는 l1 정규화를 적용함"""
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(unit, activation=tf.nn.relu, input_shape=(10000,), kernel_regularizer="l1"),
        tf.keras.layers.Dense(unit, activation=tf.nn.relu),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["acc"])
    return model
````

### [드롭아웃 추가하기](https://www.tensorflow.org/tutorials/keras/overfit_and_underfit?hl=ko#%EB%93%9C%EB%A1%AD%EC%95%84%EC%9B%83_%EC%B6%94%EA%B0%80%ED%95%98%EA%B8%B0)

학습할 때 `Dropout` 앞에 있는 레이어의 유닛을 랜덤하게 비활성화시킨다.  
한번 학습할 때 전체 가중치 중에 일부만 업데이트되기 때문에 오버피팅을 지연시킨다.  
드롭아웃을 사용하더라도 학습은 수렴한다.  
예측할 떄는 Dropout이 활성되지 않는다. 예측할 때는 모든 유닛을 사용한다.


<center>(왼쪽) 일반적인 Fully Connected  (오른쪽) Dropout을 적용하여 Fully Connected되지 않게함</center>  

<center>이미지 출처: http://blog.skby.net/드롭아웃-dropout/</center>

````python
def create_model(unit):
    """
    l1 정규화, 드롭아웃 을 적용한 모델
    
    Arguments
    ---------
        unit: 레이어의 unit 수
    """
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(unit, activation=tf.nn.relu, input_shape=(10000,), kernel_regularizer="l1"),
        tf.keras.layers.Dropout(0.2), # 랜덤하게 20%의 유닛을 비활성화
        tf.keras.layers.Dense(unit, activation=tf.nn.relu),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["acc"])
    return model
````

```
정리하면 신경망에서 과대적합을 방지하기 위해 가장 널리 사용하는 방법은 다음과 같습니다:

- 더 많은 훈련 데이터를 모읍니다.
- 네트워크의 용량을 줄입니다.
- 가중치 규제를 추가합니다.
- 드롭아웃을 추가합니다.
이 문서에서 다루지 않은 중요한 방법 두 가지는 데이터 증식(data-augmentation)과 배치 정규화(batch normalization)입니다.  
  
<출처 : Tensorflow tutorial> 
```

이 바닥은 돈이 많아야한다. unit 수를 늘려보면 레이어 3개밖에 없는데도 학습이 꽤 오래 걸린다.

````python
model1 = create_model(512)
````

````python
result = model1.fit(train_data, y_train, epochs=2, batch_size=512,
                    validation_data=(test_data, y_test), 
                    verbose=2)
````

일반적으로 validation 점수가 train 보다 클 수 없다.

### duck typing

- python은 duck typing을 지원한다.
    - 실제 오리인지 확인하지 않아도, 오리처럼 꽥꽥 울면 → 오리라고 판단한다.
    - 객체의 type을 체크하지 않고, method나 attribute가 있으면 → 사용할 수 있다. 
- `KerasClassifier`와 `KerasRegressor`는 duck typing으로 만들었다.
    - scikit-learn 알고리즘이 가지고 있는 `fit`, `predict`, `get_params` 등의 속성을 모두 가지고 있기 때문에 scikit-learn에서 그대로 사용할 수 있다.

`Dataset`을 이용하면 pandas 데이터를 tensor로 바로 바꿀 수 있다.  
데이터셋을 다루기 때문에 `shuffle`, `batch`도 지원한다.  
튜토리얼에서는 함수로 만들었다.

````python
# 튜토리얼 코드

# 판다스 데이터프레임으로부터 tf.data 데이터셋을 만들기 위한 함수
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    dataframe = dataframe.copy()
    labels = dataframe.pop('target')
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    return ds
````

다음 실습을 위해서 테스트용 샘플 데이터 하나를 준비한다.

````python
sample_data, sample_label = next(iter(df_to_dataset(data)))
````

### feature_column

`feature_column`은 pandas의 column을 keras 모델 학습에 필요한 feature로 맵핑해준다.

````python
from tensorflow import feature_column
````

age 열은 숫자 타입이다. `numeric_comlumn` 함수로 숫자형 tensor 로 변환할 수 있다.

````python
age = feature_column.numeric_column("age")
````

`DenseFeatures`는 feature_column을 일괄처리해주는 파이프라인 레이어이다.  
 input 데이터에서 key가 age인 데이터를 숫자형 tensor로 변환해준다.

````python
# keras는 functional paradigm을 지원하므로, 인스턴스에 바로 괄호를 붙일 수 있음
tf.keras.layers.DenseFeatures([age])(sample_data)
````

모델 내에서도 잘 동작한다.

````python
model = tf.keras.models.Sequential([
    tf.keras.layers.DenseFeatures([age])
])
````

````python
model(sample_data)
````

thal열은 데이터 타입이 문자열이다.  
one hot encoding으로 전처리 안해도 `feature_column`이 알아서 인코딩해준다.

````python
thal = feature_column.categorical_column_with_vocabulary_list(
      'thal', ['fixed', 'normal', 'reversible'])

# neural network는 0과 1사이로 바꾸는걸 좋아하기 때문에 onehotencoding을 더 선호한다.
thal_one_hot = feature_column.indicator_column(thal)
````

````python
# 인코딩 안해도 알아서 해준다.
tf.keras.layers.DenseFeatures([thal_one_hot])(sample_data)
````

sex열도 데이터 타입은 숫자형이지만, 실질적인 의미로는 범주형(category type)이다.  
따라서 `numeric_column` 대신 `categorical_column_with_identity`을 사용한다.

````python
sex = feature_column.categorical_column_with_identity("sex", 2)
sex_one_hot = feature_column.indicator_column(sex)
````

````python
tf.keras.layers.DenseFeatures([sex_one_hot])(sample_data)
````

이외에도 pandas로부터 다양한 데이터 타입을 tensor로 바꿔주는 기능이 있다.    
더 많은 기능은 tensorflow tutorial : [여러 종류의 특성 열 알아보기](https://www.tensorflow.org/tutorials/structured_data/feature_columns?hl=ko#%EC%97%AC%EB%9F%AC_%EC%A2%85%EB%A5%98%EC%9D%98_%ED%8A%B9%EC%84%B1_%EC%97%B4_%EC%95%8C%EC%95%84_%EB%B3%B4%EA%B8%B0)를 참고하자.

`DenseFeatures`에 feature column을 여러개 넣으면 다음과 같이 통합된 tensor로 만들어준다.  
일종의 pipeline같은 역할이다.

````python
model = tf.keras.models.Sequential([
    tf.keras.layers.DenseFeatures([age, thal_one_hot, sex_one_hot])
])
````

````python
model(sample_data)
````

**feature column의 장점**
    - 전처리 안해줘도 데이터 학습 시킬때 들어가면서 feature column으로 지정해주면 알아서 전처리 된다. 
    - feature selection도 지원한다.

기존의 pandas, scikit-learn, keras는 다음과 같은 연동 구조이다.

||||||
|---|---|---|---|---|
|pandas |--(연동)→| scikit-learn |        |       | 
|       |         | scikit-learn |--(연동)→| keras |
|pandas |          | --(연동x)→   |        | keras |

scikit-learn을 통해서 keras와 pandas를 붙일 수 있다.  
하지만 단점은, pandas와 scikit-learn은 데이터 타입이 numpy라서 tensor로 바꿀 때 loss가 생긴다.  
tensorflow 2.0 업데이트하면서 pandas를 바로 tensor로 만들어주는 기능이 추가되었다.  
scikit-learn에서 파이프라인을 만들지 않아도 keras에서 바로 pandas 데이터 사용이 가능해졌다.
