#### StandardScaler

${x-\mu}\over{\sigma}$ ($\mu$:평균, $\sigma$:표준편차)
- 표준정규분포로 변환한다.
- scaling 후 평균은 0, 표준편차는 1이 된다.

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

scikit-learn은 쓰기 정말 쉽다. 이번에도 인스턴스 -> fit 한다.  
단, `StandardScaler`는 shape을 잘 맞춰야한다.

```python
# Error 
# DataFrame에 .으로 접근하면 Series(1차원)이다.
ss.fit(wine.proline)
```

```python
# fancy indexing으로 뽑으면 DataFrame(2차원)
ss.fit(wine[["proline"]])
```

`predict` 대신 preprocessing에 있는 애들은 `transfrom`한다.

```python
ss.transform(wine[["proline"]])
```

`fit_transform`은 `fit`과 `transfrom`을 한번에 해주는 메소드이다.

```python
ss.fit_transform(wine[["proline"]])
```

재할당해서 정규화한 값으로 변경한다.

```python
wine.proline = ss.fit_transform(wine[["proline"]])
```

스케일링 후에 `boxplot`을 다시 해본다.

#### Dummy Variable / OneHotEncoding

남자는 [0, 1]
여자는 [1, 0] 으로 바꾸는 것을 통계 용어로 dummy 변수라고 한다.
  
CS쪽에서는 OneHotEncoding이라고 부른다.

```python
pd.get_dummies(tips.sex) # 열이 1개에서 2개로 바뀜
```

day를 원핫인코딩해보자.  
카테고리의 개수만큼 열(column)이 생긴다.

```python
pd.get_dummies(tips.day)
```

문자를 숫자로 바꾸는 것을 encoding:
    - Label Encoding
    - OneHot Encoding (1이 1개인 것)
    - MultiHot Encoding (1이 여러개인 것) -> 우리의 관심사 아님
상황에 따라서 적절한 encoding 방법을 선택해야 한다.

### missing data 채우기

`isnull`메서드로 pandas에서 missing data의 위치를 찾을 수 있다.

```python
mpg.isnull()
```

```python
mpg[mpg.horsepower.isnull()] # boolean indexing으로 missing 있는 행만 가져오기
```

```python
mpg[mpg.horsepower.isnull()].fillna(mpg.horsepower.mean()) # 평균값 집어넣기
```

```python
mpg[mpg.horsepower.isnull()].fillna(mpg.horsepower.median()) # 중앙값 집어넣기 
```

또 다른 전략:  
빈도를 기반으로 class를 나눠서 categorical data로 만든다.  
이 방법에는 `map`을 활용할 수 있다.

```python
def name(x):
    if x == "ford pinto":
        return 6
    elif x == "amc matador":
        return 5
    else:
        return 0
```

```python
mpg.name.map(name)
```
