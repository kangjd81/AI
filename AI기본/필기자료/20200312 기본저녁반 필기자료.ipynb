matplotlib은 matlab이라는 프로그램을 베껴서 만든 시각화 라이브러리이다.  
처음에는 `from pylab import *` 방식으로 임포트하여, matlab과 동일한 방법으로 사용했다.

현재는 pyplot을 임포트해서 사용하는 방식이 권장된다.

matplotlib은 3가지 방식으로 그래프를 그릴 수 있다.
- State Machine
- Objective Oriented Programming

### State Machine 방식

figure는 가상의 그래프 그릴 공간이다.

````python
plt.figure()
````

axis는 그래프이다.  
figure만 했을 때와 달리 무언가 생겼다.

````python
plt.figure()
plt.axis()
````

`figure`에 `figsize`옵션을 주면 그래프가 커진다.

````python
plt.figure(figsize=(10,5)) 
plt.axis() 
````

````python
plt.axes()
````

State Machine은 특별히 지정하지 않아도, 같은 상태를 공유하는 방식이다.  
앞서 `figure`를 실행한 후, `axis`를 실행해서 figure 내에 그래프(axis)가 그려지도록 했다.  
그러나, `axis`를 실행할 때 figure에 대한 어떤 정보도 제공하지 않았는데 자동으로 `figure`와 연동하였다.  
실제로 `figure`에 `figsize` 옵션을 넣었을 때, `axis`의 결과에도 영향을 미쳤었다. 

jupyter notebook에서는 하나의 셀은 같은 상태를 공유한다.  
그러므로, matplotlib을 사용할 때는 한 셀 안에 코드를 모두 써야한다.

````python
plt.figure(figsize=(10,5))
````

````python
plt.axes() # 셀을 나눠서 실행하면 안 커진다.
````

`figure`가 그래프의 가장 기본이므로 반드시 있어야하는데,

### Objective Oriented Programming방식

`subplots`는 객체지향 방식으로 그래프를 그린다.  
위에서 언급한 `subplot`에 s 한글자가 더 붙었다.
`subplots`는 figure, axes 객체를 반환한다.

````python
plt.subplots()
````

````python
fig, axes = plt.subplots() # unpacking 할당 
````

객체 지향 프로그래밍처럼 axes 객체에 .으로 접근해서 속성을 변경한다.  
※ 참고
- 원래는 fig.show()를 해야 그래프가 보인다. 그러나 jupyter notebook은 fig.show() 안해도 된다. 
    (다른 파이썬 개발환경에서는 fig.show()를 해야한다.)  
- 객체 지향 방식으로 그래프를 그릴 때에도, 같은 셀 안에 코드가 모두 있어야한다.

````python
fig, axes = plt.subplots()
axes.set_title("title")
# fig.show()
````

nrows, ncols 파라미터로 전체 그래프 배열을 정한다.  
그래프를 한 개 이상 할당하면 axes가 array로 나온다. (리턴 확인)

````python
plt.subplots(1, 3) # 가로 1줄, 세로 3줄
````

array로 되어 있는 axes는 인덱싱으로 뽑아낼 수 있다.  
뽑아낸 axes에 .으로 접근해서 해당 그래프만 조절할 수 있다.

````python
fig, axes = plt.subplots(1,3)
axes[0].set_title("1")
axes[2].set_title("2")
````

디자인이 조금 나아졌지만, 와이프도 공돌이 감성이었다.  
matplotlib의 원 개발자 John D hunter가 타계한 후 스타일 기능이 추가되었다.

## 한글 폰트 설정

````python
import matplotlib.pyplot as plt
````

font_manager는 matplotlib에서 폰트를 관장하는 모듈이다.

````python
from matplotlib import font_manager
````

`FontManager`는 camel 방식의 이름인 것을 보고 클래스임을 유추할 수 있다.  
클래스는 인스턴스화해서 사용한다.

````python
fm = font_manager.FontManager()
````

````python
dir(fm) # 처음 볼 때는 dir
````

`ttffiles`는 사용가능한 폰트 파일의 주소를 반환한다.

````python
fm.ttffiles # PC에 설치된 폰트에 따라서 결과가 다름. 
            # 너무 많아서 일부만 출력하였음
````

`ttflist`는 사용가능한 폰트 정보를 반환한다.

````python
fm.ttflist  # PC에 설치된 폰트에 따라서 결과가 다름. 
            # 너무 많아서 일부만 출력하였음
````

`ttffiles`, `ttflist`에서 한글 지원이 되는 폰트 이름을 찾아서 `rcParmams`의 "font.family"를 재할당한다.  
PC마다 폰트가 다르기 때문에 한글지원 폰트는 직접 찾아야 한다.

- 나눔 폰트 설치했을 경우 나눔(Nanum)은 한글 가능

### 단계구분도 (Choropleth)

단계 구분도(Choropleth)는 구역별로 색을 칠하는 것이다.  
색칠의 기준이 되는 데이터는 pandas와 연동한다.  
먼저, 연동할 pandas 데이터를 만들어보자.

````python
data = pd.DataFrame.from_dict(seoul_geo_json["features"])
````

````python
data
````

`map`은 아주 편리한 기능이다.  
map을 이용하면 위 DataFrame에서 경도, 위도만 뽑아오는 것도 가능하다.

````python
t = data.geometry.map(lambda x:x["coordinates"][0][0]) # 경도, 위도 뽑기
````

````python
t
````

map을 이용해서 단계구분도에 사용한 DataFrame을 만들자.

````python
data_list = {}
data_list["base_year"] = data.properties.map(lambda x: x["base_year"])
data_list["code"] = data.properties.map(lambda x: x["code"])
data_list["name"] = data.properties.map(lambda x: x["name"])
data_list["name_eng"] = data.properties.map(lambda x: x["name_eng"])
````

````python
seoul_df = pd.DataFrame.from_dict(data_list)
seoul_df = seoul_df.astype({"base_year":"int64", "code":"int64"})
````

만들어진 DataFrame은 to_로 시작하는 메소드로 저장할 수도 있다.

````python
seoul_df.to_csv("seoul_geo_df.csv")
````

단계 구분도는 다음과 같이 그릴 수 있다.

````python
m9 = folium.Map([37.5665, 126.9780])

folium.Choropleth(seoul_geo_json, 
                  seoul_df, 
                  ["name", "code"], 
                  key_on="feature.properties.name", # features 아님 주의
                  fill_color="YlGn", 
                  nan_fill_color="black"
                 ).add_to(m9)
````
